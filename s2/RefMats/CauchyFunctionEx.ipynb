```python
# Example of plotting a cauchy distribution 

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import cauchy

# Parameters
x0 = 0  # location parameter
gamma = 1  # scale parameter

# Generate x values
x = np.linspace(-10, 10, 1000)

# Compute the PDF
pdf = cauchy.pdf(x, x0, gamma)

# Plot the PDF
plt.plot(x, pdf, label='Cauchy Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Cauchy Distribution (x0=0, gamma=1)')
plt.legend()
plt.show()
```
