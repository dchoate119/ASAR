## Step 2: Calculate Gradient Images 

```python
# Choose specific image 
image_fnc = "/home/daniel-choate/ASAR/s2/TerrainNav/Photo_tests/tuftsww_colm_resized.png"
image_fnge = "/home/daniel-choate/ASAR/s2/TerrainNav/Photo_tests/tuftsww_ge_resized.png"

imc = cv2.imread(image_fnc)
imge = cv2.imread(image_fnge)

imc = np.float32(imc) / 255.0
imge = np.float32(imge) / 255.0

imc = cv2.cvtColor(imc, cv2.COLOR_BGR2RGB) # Open CV uses BGR??
imge = cv2.cvtColor(imge, cv2.COLOR_BGR2RGB) # Open CV uses BGR??

print(imc.shape)
print(imge.shape)

# Calculate gradient
gx_c = cv2.Sobel(imc, cv2.CV_32F, 1, 0, ksize=1)
gy_c = cv2.Sobel(imc, cv2.CV_32F, 0, 1, ksize=1)

gx_ge = cv2.Sobel(imge, cv2.CV_32F, 1, 0, ksize=1)
gy_ge = cv2.Sobel(imge, cv2.CV_32F, 0, 1, ksize=1)

print(gx_c.shape)
print(gx_ge.shape)

# print("\n Gradients in the x-direction, colm \n", gx_c)
# print("\n Gradients in the y-direction, colm \n", gy_c)
# print("\n Gradients in the x-direction, ge \n", gx_ge)
# print("\n Gradients in the y-direction, ge \n", gy_ge)


# Calculate gradient magnitude and direction ( in degrees )
mag_c, angle_c = cv2.cartToPolar(gx_c, gy_c, angleInDegrees=True)
mag_ge, angle_ge = cv2.cartToPolar(gx_ge, gy_ge, angleInDegrees=True)


# Make ALL ANGLES within 180 degrees 
angle_c[angle_c > 180] -= 180
angle_ge[angle_ge > 180] -= 180

print(mag_c.shape)
print(mag_ge.shape)

print("\n Angles colm \n", angle_c)
print("\n Angles ge \n", angle_ge)
```

## Step 3: Divide Image into 8x8 cells

```python
# First: for color data 
# Need maximum magnitude
# Need angle associated with that maximum magnitude 

# COLMAP
mag_max_ind = np.argmax(mag_c, axis=2)
# print(mag_max_ind.shape)
mag_simp = np.max(mag_c, axis=2)
# print(mag_simp.shape)
mag_simp = np.expand_dims(mag_simp, axis=2)
# print(mag_simp.shape)
ang_simp = np.take_along_axis(angle_c, mag_max_ind[..., np.newaxis], axis=2)
# Combine magnitude and angle into one array 
magang_colm = np.concatenate((mag_simp, ang_simp), axis=2)
print(magang_colm.shape)

# Google Earth 
mag_max_ind = np.argmax(mag_ge, axis=2)
# print(mag_max_ind.shape)
mag_simp = np.max(mag_ge, axis=2)
# print(mag_simp.shape)
mag_simp = np.expand_dims(mag_simp, axis=2)
# print(mag_simp.shape)
ang_simp = np.take_along_axis(angle_ge, mag_max_ind[..., np.newaxis], axis=2)
# Combine magnitude and angle into one array 
magang_ge = np.concatenate((mag_simp, ang_simp), axis=2)
print(magang_ge.shape)
```

```python
def divide_into_blocks(im_arr, block_size=8):
    """ Takes an array of pixels containing a magnitude and angle, and
    divides them into blocks, size 8x8
    Inputs: image_array
    Outputs: blocks of 8x8 pixels
    """
    # Get the dimensions of the array (n, m, 2)
    n, m, c = im_arr.shape

    # Ensure that n is twice the size of m
    if n != 2 * m or c != 2:
        raise ValueError("Array must have shape (2*m, m, 2)")

    # Ensure that both n and m are divisible by the block size
    if n % block_size != 0 or m % block_size != 0:
        raise ValueError("Array dimensions must be divisible by the block size")

    # Number of blocks along each axis
    num_blocks_n = n // block_size
    num_blocks_m = m // block_size

    # Reshape and swap axes to divide into blocks of shape (block_size, block_size, 2)
    blocks = im_arr.reshape(num_blocks_n, num_blocks_m, block_size, block_size, 2).swapaxes(1, 2)

    return blocks

cells_colm = divide_into_blocks(magang_colm)
cells_ge = divide_into_blocks(magang_ge)

print("Colmap image divided blocks shape:", cells_colm.shape)  # Should be (16, 8, 8, 8, 2) for this example
print("Google earth image divided blocks shape:", cells_ge.shape)  # Should be (16, 8, 8, 8, 2) for this example

print("\n Colmap divided blocks: \n", cells_colm)  
print("\n Google Earth divided blocks: \n", cells_ge)  
```

## Step 4: Compute histogram of gradients for each cell

```python
# Using proportionate version of histogram distribution 
# See HOG_trial, for other versions

def compute_hog_for_cells(mag_angle_array, num_bins=9):
    # Get the number of cells in each dimension
    num_cells_x, num_cells_y, cell_size_x, cell_size_y, _ = mag_angle_array.shape
    
    # Define the bin edges (0 to 180 degrees for orientations)
    bin_edges = np.linspace(0, 180, num_bins + 1)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
        
    # Initialize an empty array to hold the histograms for each cell
    histograms = np.zeros((num_cells_x, num_cells_y, num_bins))
    
    # Loop over each cell
    for i in range(num_cells_x):
        for j in range(num_cells_y):
            # Extract the magnitudes and angles for this 8x8 cell
            cell_magnitudes = mag_angle_array[i, j, :, :, 0]
            cell_angles = mag_angle_array[i, j, :, :, 1]
            
            # Initialize histogram for this cell
            histogram = np.zeros(num_bins)

            # Loop over each pixel in the cell
            for x in range(cell_size_x):
                for y in range(cell_size_y):
                    mag = cell_magnitudes[x, y]
                    # print(mag)
                    angle = cell_angles[x, y]
                    # print(angle)
                    
                    # Assign the angle to a bin
                    bin_idx = np.digitize(angle, bin_edges) - 1  # Find bin index
                    # print(bin_idx)
                    bin_idx = min(bin_idx, num_bins - 1)  # Ensure it stays within bounds
                    # print(bin_idx)

                    if angle == 0 or bin_idx == num_bins - 1:
                        # If the angle is at the boundary, assign it fully to the closest bin
                        histogram[bin_idx] += mag
                    else:
                        # Get the bin edges for interpolation
                        lower_edge = bin_edges[bin_idx]
                        upper_edge = bin_edges[bin_idx + 1]

                        # Calculate proportion of magnitude for each bin
                        ratio_upper = (angle - lower_edge) / (upper_edge - lower_edge)
                        # print(ratio_upper)
                        ratio_lower = 1 - ratio_upper

                        # Distribute magnitude proportionally
                        histogram[bin_idx] += mag * ratio_lower
                        histogram[bin_idx + 1] += mag * ratio_upper

            
            # Store the histogram for the current cell
            histograms[i, j, :] = histogram

    return histograms

# Assume mag_angle_array has the shape (16, 8, 8, 8, 2)
# First 4 dimensions are the grid of cells and pixels within the cells,
# the last dimension contains magnitude and angle.
# If number of bins is 9, function results in a 9x1 array input into each cell (16x8x9)

hog_histograms_c = compute_hog_for_cells(cells_colm)
hog_histograms_ge = compute_hog_for_cells(cells_ge)

print("Colmap HOG Histograms for each cell: \n", hog_histograms_c.shape)
print("Google Earth HOG Histograms for each cell: \n", hog_histograms_ge.shape)

# WE NOW HAVE A 9-DIM 'FEATURE VECTOR' FOR EACH PIXEL 
```

```python
def plot_hog_histogram(hist, bin_edges, cell_position=None):
    """
    Plots the HOG histogram for a single cell.
    
    Parameters:
    hist (array): The histogram of gradient orientations for the cell.
    bin_edges (array): The edges of the orientation bins (angles).
    cell_position (tuple): The (i, j) position of the cell, optional, for labeling.
    """
    # Define the bin centers (midpoints between bin edges)
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    
    # Plot histogram as a bar chart
    plt.bar(bin_centers, hist, width=(bin_edges[1] - bin_edges[0]), align='center', edgecolor='black')
    
    # Set title and labels
    if cell_position is not None:
        plt.title(f'HOG Histogram for Cell {cell_position}')
    else:
        plt.title('HOG Histogram')
    plt.xlabel('Orientation (degrees)')
    plt.ylabel('Magnitude')
    plt.grid(True)

# Example of how to plot HOG histograms for one or more cells
num_bins = 9
bin_edges = np.linspace(0, 180, num_bins + 1)

# Assuming hog_histograms is a (16, 8, num_bins) array (cell histograms)
cell_i, cell_j = 3, 3  # Choose the cell (0, 0) for example
histc = hog_histograms_c[cell_i, cell_j, :]  # Extract the histogram for the selected cell
histge = hog_histograms_ge[cell_i, cell_j, :]  # Extract the histogram for the selected cell

# Plot the histogram for the selected cell
plt.figure(figsize=(6, 4))
plt.subplot(1, 2, 1)
plot_hog_histogram(histc, bin_edges, cell_position=(cell_i, cell_j))

plt.subplot(1, 2, 2)
plot_hog_histogram(histge, bin_edges, cell_position=(cell_i, cell_j))

plt.show()
```

## Step 5: Club blocks in groups of four

```python
# COLMAP
epsilon = 1e-05
feature_vectors_colm = []
for i in range(0, len(hog_histograms_c) - 1, 1):
    temp = []
    for j in range(0, len(hog_histograms_c[0]) - 1, 1):
        values_c = [[hog_histograms_c[i][x] for x in range(j, j+2)] for i in range(i, i+2)]
        # print("Values are \n",values)
        final_vector = []
        for k in values_c:
            for l in k:
                for m in l:
                    # print(m)
                    final_vector.append(m)
        # print(final_vector, i)
        k = round(math.sqrt(sum([pow(x, 2) for x in final_vector])), 9)
        final_vector = [round(x/(k + epsilon), 9) for x in final_vector]
        # print(temp)
        temp.append(final_vector)
    feature_vectors_colm.append(temp)

# GOOGLE EARTH
epsilon = 1e-05
feature_vectors_ge = []
for i in range(0, len(hog_histograms_ge) - 1, 1):
    temp = []
    for j in range(0, len(hog_histograms_ge[0]) - 1, 1):
        values_ge = [[hog_histograms_ge[i][x] for x in range(j, j+2)] for i in range(i, i+2)]
        # print("Values are \n",values)
        final_vector = []
        for k in values_ge:
            for l in k:
                for m in l:
                    # print(m)
                    final_vector.append(m)
        # print(final_vector, i)
        k = round(math.sqrt(sum([pow(x, 2) for x in final_vector])), 9)
        final_vector = [round(x/(k + epsilon), 9) for x in final_vector]
        # print(temp)
        temp.append(final_vector)
    feature_vectors_ge.append(temp)

```

## Step 6: Plot histograms over image 

```python
def plot_hog_on_image(image, hog_histograms, cell_size=8, num_bins=9):
    # Display the original image
    plt.imshow(image, cmap="gray")
    
    # Get image dimensions
    img_height, img_width = image.shape[:2]
    
    # Create a mesh grid for the positions of the cell centers
    num_cells_y, num_cells_x = hog_histograms.shape[:2]
    print(num_cells_y, num_cells_x)
    x = np.linspace(cell_size / 2, img_width - cell_size / 2, num_cells_x)
    y = np.linspace(cell_size / 2, img_height - cell_size / 2, num_cells_y)
    X, Y = np.meshgrid(x, y)
    
    # Prepare for plotting arrows (gradients)
    angles_per_bin = 180 / num_bins
    for i in range(num_cells_y):
        for j in range(num_cells_x):
            histogram = hog_histograms[i, j]
            
            for bin_idx in range(num_bins):
                # Get the magnitude and angle for this bin
                magnitude = histogram[bin_idx]
                angle = bin_idx * angles_per_bin + angles_per_bin / 2  # Midpoint of bin
                
                # Convert angle to radians for plotting
                angle_rad = np.deg2rad(angle)
                
                # Compute the x and y components of the vector (arrow)
                dx = magnitude * np.cos(angle_rad)
                dy = magnitude * np.sin(angle_rad)
                
                # Plot the arrow
                plt.quiver(X[i, j], Y[i, j], dx, dy, color='white', angles='xy', scale_units='xy', scale=1)
    
    # Display the combined image with HOG arrows
    plt.axis("off")
    # plt.show()

# Example Usage:
# Colmap 
image_fn = "/home/daniel-choate/ASAR/s2/TerrainNav/Photo_tests/tuftsww_colm_resized.png"
image_c = cv2.imread(image_fn)#, cv2.IMREAD_GRAYSCALE)  
image_c = cv2.cvtColor(image_c, cv2.COLOR_BGR2RGB) # Open CV uses BGR??
image_c = np.float32(image_c) / 255.0  # Normalize

# Google Earth 
image_fn = "/home/daniel-choate/ASAR/s2/TerrainNav/Photo_tests/tuftsww_ge_resized.png"
image_ge = cv2.imread(image_fn)#, cv2.IMREAD_GRAYSCALE)  
image_ge = cv2.cvtColor(image_ge, cv2.COLOR_BGR2RGB) # Open CV uses BGR??
image_ge = np.float32(image_ge) / 255.0  # Normalize

# Example HOG histograms: Assuming you already have `hog_histograms` computed
# hog_histograms = compute_hog_for_cells(mag_angle_array)
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plot_hog_on_image(image_c, hog_histograms_c)

plt.subplot(1, 2, 2)
plot_hog_on_image(image_ge, hog_histograms_ge)

plt.show()
```
