## Step 2: Grab unit vector based on pixel location

The vector $u_x$ is defined as a 3x1 matrix:

$$
\mathbf{u_g} = \begin{bmatrix} p_x \\ p_y \\ f \end{bmatrix}
$$

Where:
- \(p_x\) represents the x-coordinate,
- \(p_y\) represents the y-coordinate,
- \(f\) represents focal length.

NOTE: for a unit vector, divide by magnitude 

NOTE: point location must be shifted to the middle of the model

## Step 3: Change unit vector into world coords 

```python
# Define necessary functions 
# Get camera pose for chosen image 
def get_pose_id(id):
    qvec = images_colmap[id].qvec
    tvec = images_colmap[id].tvec[:,None]
    # print(tvec)
    
    t = tvec.reshape([3,1])
    # print(tvec)
    Rotmat = qvec2rotmat(qvec) # Positive or negative does not matter
    # print("\n Rotation matrix \n", Rotmat)
    
    bottom = np.array([0.0, 0.0, 0.0, 1.0]).reshape([1, 4])
    m = np.concatenate([np.concatenate([Rotmat, t], 1), bottom], 0)
    c2w = np.linalg.inv(m)
    return c2w

def inv_homog_transform(homog):
    """ Inverting a homogeneous transformation matrix
    Inputs: homogeneous transformation matrix (4x4)
    Outputs: inverted 4x4 matrix
    """
    # Grab rotation matrix
    R = homog[:3,:3]
    # print("\n R \n",R)
    # Transpose rotation matrix 
    R_inv = R.T
    # print("\n Transpose of R \n",R_inv)
    # Grab translation matrix 
    t = homog[:-1, -1]
    t = t.reshape((3, 1))
    # print("\n T \n",t)
    # - R_t * t
    t_inv = -R_inv @ t
    # print("\n -R_t * T \n",t_inv)
    bottom = np.array([0.0, 0.0, 0.0, 1.0]).reshape([1, 4])
    homog_inv = np.concatenate([np.concatenate([R_inv, t_inv], 1), bottom], 0)    
    # print("\n Homogeneous new = \n", homog_inv)
    return homog_inv
```

```python
# Get transformation matrix that move from camera coords to world coords 
homog_c2w = get_pose_id(im_ID)
print('Homogeneous transformation from world to camera \n', homog_c2w)

# Inverse for world coords to camera coords
homog_w2c = inv_homog_transform(homog_c2w)
print('\n Homogeneous transformation from camera to world \n', homog_w2c)
```

## 5a: Using SVD for gravity vector 

```python
pts_gnd = np.array([scene_pts[24140,:],scene_pts[24142,:], scene_pts[24146]])
print("Original points \n",pts_gnd_T)
centroid = np.mean(pts_gnd_T, axis=0)
# Center the points
centered_points = pts_gnd_T - centroid

# Compute SVD
print("Centered points \n", centered_points)

U, S, Vt = np.linalg.svd(centered_points)

# Display results
print("U matrix:\n", U)
print("Singular values:\n", S)
print("Vt matrix:\n", Vt)

# new_grav = Vt[:,-1]
new_grav = Vt[-1, :]
print("New gravity vector\n", new_grav)

dot_product = np.dot(grav_dir, new_grav)
if dot_product < 0:
    new_grav = -new_grav  # Flip direction if necessary
print("New gravity vector\n", new_grav)
```
