```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

def read_colmap_depth(path):
    with open(path, 'rb') as f:
        raw = f.read()

    import re
    header_match = re.match(rb'(\d+&\d+&\d+&)', raw)
    if not header_match:
        raise ValueError("Header not found in expected format")
    header_bytes = header_match.group(1)
    header_str = header_bytes.decode('ascii').strip('&')
    width, height, channels = map(int, header_str.split('&'))
    print(f"Parsed width={width}, height={height}, channels={channels}")
    assert channels == 1, "This function is for depth maps (1 channel) only"

    # Skip null bytes after header
    header_end = len(header_bytes)
    while header_end < len(raw) and raw[header_end] == 0:
        header_end += 1

    float_data_bytes = raw[header_end:]
    actual_count = len(float_data_bytes) // 4
    max_valid_height = actual_count // width

    if max_valid_height < height:
        print(f"Warning: Truncating height from {height} to {max_valid_height} due to short file")
        height = max_valid_height

    float_data_bytes = float_data_bytes[:height * width * 4]
    data = np.frombuffer(float_data_bytes, dtype=np.float32).copy()
    data = data.reshape((height, width))
    return data


def read_colmap_normal(path):
    with open(path, 'rb') as f:
        raw = f.read()

    import re
    header_match = re.match(rb'(\d+&\d+&\d+&)', raw)
    if not header_match:
        raise ValueError("Header not found in expected format")
    header_bytes = header_match.group(1)
    header_str = header_bytes.decode('ascii').strip('&')
    width, height, channels = map(int, header_str.split('&'))
    print(f"Parsed width={width}, height={height}, channels={channels}")
    assert channels == 3, "This function is for normal maps (3 channels) only"

    header_end = len(header_bytes)
    while header_end < len(raw) and raw[header_end] == 0:
        header_end += 1

    float_data_bytes = raw[header_end:]
    actual_count = len(float_data_bytes) // 4
    max_valid_height = actual_count // (width * channels)

    if max_valid_height < height:
        print(f"Warning: Truncating height from {height} to {max_valid_height} due to short file")
        height = max_valid_height

    float_data_bytes = float_data_bytes[:height * width * channels * 4]
    data = np.frombuffer(float_data_bytes, dtype=np.float32).copy()
    data = data.reshape((height, width, channels))
    return data
```

```python
# depth_path = '/home/daniel-choate/Datasets/COLMAP/GerrardHall/dense/0/stereo/depth_maps/IMG_2331.JPG.geometric.bin'
depth_path = '/home/daniel-choate/Datasets/COLMAP/TTurfSAT/dense/0/stereo/depth_maps/IMG_9446.JPEG.photometric.bin'
depth_map = read_colmap_depth(depth_path)

depth_map[depth_map == 0] = np.nan
plt.imshow(depth_map, cmap='plasma')
plt.colorbar(label='Depth (m)')
plt.title('COLMAP Depth Map')
plt.show()
```

```python
# normal_path = '/home/daniel-choate/Datasets/COLMAP/GerrardHall/dense/0/stereo/normal_maps/IMG_2331.JPG.geometric.bin'
normal_path = '/home/daniel-choate/Datasets/COLMAP/TTurfSAT/dense/0/stereo/normal_maps/IMG_9446.JPEG.photometric.bin'
normal_map = read_colmap_normal(normal_path)

# Normalize from [-1, 1] to [0, 1] for RGB-like display
normal_viz = (normal_map + 1) / 2.0
plt.imshow(normal_viz)
plt.title("COLMAP Normal Map")
plt.axis('off')
plt.show()
```
