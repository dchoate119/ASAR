```python
import numpy as np
import pandas as pd
import cv2
import open3d as o3d
import matplotlib.pyplot as plt

from pathlib import Path

# SAVE YOUR WORK
%load_ext autoreload
%autoreload 2
%autosave 180
```

```python
# Import data
imu_path = '/home/daniel-choate/Datasets/OU_SurfNav/Etracted_Data/imu_data.txt'
img_path = '/home/daniel-choate/Datasets/image_frames'
# Load the imu data
imu_df = pd.read_csv(imu_path, sep="\t")

# Load the image data
# img_df = pd.read_csv(img_path, sep="\n")

# Preview
print(imu_df.head())
# print(img_df.head())
```

```python
# Plot linear acceleration as a function of time

plt.figure(figsize=(10, 6))

plt.plot(imu_df['timestamp_sec'], imu_df['linear_acceleration_x'], label='X-axis')
plt.plot(imu_df['timestamp_sec'], imu_df['linear_acceleration_y'], label='Y-axis')
plt.plot(imu_df['timestamp_sec'], imu_df['linear_acceleration_z'], label='Z-axis')

plt.xlabel('Time (s)')
plt.ylabel('Linear Acceleration (m/s)')
plt.title('Linear Acceleration vs Time')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
print(imu_df['timestamp_sec'])
```

```python
# Filter the DataFrame for the desired time range
start_time = 1745000923
end_time = 1745000937
# end_time = 1745000939
filtered_df = imu_df[(imu_df['timestamp_sec'] >= start_time) & (imu_df['timestamp_sec'] <= end_time)]

# Plot linear acceleration as a function of time for the filtered data
plt.figure(figsize=(10, 6))

plt.plot(filtered_df['timestamp_sec'], filtered_df['linear_acceleration_x'], label='X-axis')
plt.plot(filtered_df['timestamp_sec'], filtered_df['linear_acceleration_y'], label='Y-axis')
plt.plot(filtered_df['timestamp_sec'], filtered_df['linear_acceleration_z'], label='Z-axis')

plt.xlabel('Time (s)')
plt.ylabel('Linear Acceleration (m/s)')
plt.title(f'Linear Acceleration vs Time ({start_time} to {end_time})')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
