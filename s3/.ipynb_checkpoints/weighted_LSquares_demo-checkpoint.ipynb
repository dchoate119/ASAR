```python
# Generate projection of image sections 
for i in range(len(gnav.images_dict)):
    # STEP 1: Unit vectors in camera coords 
    pts_vec_c, pts_rgb_gnd = gnav.unit_vec_c(i)
    gnav.im_mosaic[i] = {'rgbc': pts_rgb_gnd}

    # STEP 2: Transformation matrix moves from camera coords to world coords
    id = gnav.im_ids[i]
    homog_w2c, homog_c2w = gnav.get_pose_id(id,i)
    # print('Homogeneous transformation from world to camera \n', homog_c2w)
    # print('\n Homogeneous transformation from camera to world \n', homog_w2c)

    # STEP 3: Transform to world coords
    origin_c, pts_loc_w, pts_vec_w = gnav.unit_vec_tform(pts_vec_c, gnav.origin_w, homog_c2w)
    # print('\n New camera frame origin = ', origin_c)
    
    # STEP 4: Get new points 
    ranges, new_pts_w = gnav.pt_range(pts_vec_w, homog_c2w, origin_c, i)
    # print('\nNew Points \n', new_pts_w)

    # STEP 5: Transfer points to reference frame
    __, new_pts_r, pts_vec_r = gnav.unit_vec_tform(new_pts_w, gnav.origin_w, tform_ref_frame_pts)

    # Convert points to grayscale 
    gray_c = gnav.conv_to_gray(gnav.im_mosaic[i]['rgbc'],i)
    # print(gray_c)

    # Put new points and grayscale colors in image mosaic
    gnav.im_mosaic[i]['pts'] = new_pts_r
    gnav.im_mosaic[i]['color_g'] = gray_c
    
    print("\nDone image ", i)
```

### Divide each patch into micro-patches

```python
# Check number of points per patch
# Patch 3: B in jumbos
p = 2

bins_2d = {}  # store bins as {(r, c): points}

for imnum in range(len(gnav.images_dict)):
    
    # Starting with 10 micropatches
    # Grab full params
    x, y, width, height = mosaic_params[imnum]
    # print(x, y, width, height)
    
    # Bin grid
    n_rows, n_cols = 4, 4
    bin_w = width / n_cols
    bin_h = height / n_rows
    # print(bin_w, bin_h)
    
    pts = gnav.im_pts_2d[imnum]['pts']
    
    
    for r in range(n_rows):
        for c in range(n_cols):
            x_min = x + c * bin_w
            x_max = x + (c + 1) * bin_w
            y_min = y + r * bin_h
            y_max = y + (r + 1) * bin_h
    
            mask = (
                (pts[..., 0] >= x_min) & (pts[..., 0] < x_max) &
                (pts[..., 1] >= y_min) & (pts[..., 1] < y_max)
            )
            pts_bin = pts[mask]
            bins_2d[(imnum, r, c)] = pts_bin

gnav.bins_2d = bins_2d
# # Example: print number of points per bin
# for k, v in bins_2d.items():
#     print(f"Bin {k}: {v.shape[0]} points")
```

```python
# Plot to visualize bin distribution
plt.figure(figsize=(15,8))
rows = int(len(gnav.images_dict)/5)
for imnum in range(len(gnav.images_dict)):
    plt.subplot(rows,5,imnum+1)
    im = cv2.cvtColor(gnav.images_dict[imnum], cv2.COLOR_BGR2RGB)
    plt.imshow(im)
    colors = ['r', 'b', 'g', 'r', 'b', 'g', 'r', 'b', 'g', 'r', 'b', 'g', 'r', 'b', 'g', 'r']
    for r in range(n_rows):
        for c in range(n_cols):
            xs = bins_2d[(imnum,r,c)][:,0]
            ys = bins_2d[(imnum,r,c)][:,1]
            plt.plot(xs, ys, colors[4*r + c])
            plt.axis("off")
    
plt.show()
```

```python
# Transform from bins in 2D to bins in 3D
gnav.im_mosaic_micro = {}
gnav.im_pts_best_guess_micro = {}
p = 2
# Testing first for a single set of micropatches
for imnum in range(len(gnav.images_dict)):
    for b, pts in bins_2d.items():
        if b[0] != imnum:
            continue
        # STEP 1: Unit vectors in camera coords
        pts_vec_c, pts_rgb_gnd = gnav.unit_vec_c_MICRO(imnum, pts)

        # Initialize the dictionary for this image if not yet created
        if imnum not in gnav.im_mosaic_micro:
            gnav.im_mosaic_micro[imnum] = {}
        
        gnav.im_mosaic_micro[imnum][b] = {'rgbc': pts_rgb_gnd} 
    
        # STEP 2: Transformation matrix moves from camera coords to world coords
        im_id = gnav.im_ids[imnum]
        homog_w2c, homog_c2w = gnav.get_pose_id(im_id,imnum)
        # print('Homogeneous transformation from world to camera \n', homog_c2w)
        # print('\n Homogeneous transformation from camera to world \n', homog_w2c)
    
        # STEP 3: Transform to world coords 
        origin_c, pts_loc_w, pts_vec_w = gnav.unit_vec_tform(pts_vec_c, gnav.origin_w, homog_c2w)
        # print('\n New camera frame origin = ', origin_c)
    
        # STEP 4: Get new points 
        ranges, new_pts_w = gnav.pt_range(pts_vec_w, homog_c2w, origin_c, imnum)
        # print('\nNew Points \n', new_pts_w)
    
        # STEP 5: Transfer points to reference frame
        __, new_pts_r, pts_vec_r = gnav.unit_vec_tform(new_pts_w, gnav.origin_w, tform_ref_frame_pts)
    
        # Convert points to grayscale 
        gray_c = gnav.conv_to_gray(gnav.im_mosaic_micro[imnum][b]['rgbc'],imnum)
        # print(gray_c)
    
        # Put new points and grayscale colors in image mosaic
        gnav.im_mosaic_micro[imnum][b]['pts'] = new_pts_r
        gnav.im_mosaic_micro[imnum][b]['color_g'] = gray_c
        
        # print("Done patch ", b)
    print("Done image ", imnum)

# Implement best guess 
gnav.implement_guess_micro(gnav.best_guess_tform, gnav.best_guess_scale)
```
